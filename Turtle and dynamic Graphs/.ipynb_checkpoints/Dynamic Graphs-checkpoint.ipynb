## Imports & pre-requisites

```python
%matplotlib inline
%reload_ext autoreload
%autoreload 2
%matplotlib notebook
import random
import scipy.stats as stats
from matplotlib import pyplot
import numpy as np
import time
```

## 10x4 matrix creation

```python
randoms = [[random.random() for i in range(40)] for j in range(100)]
```

## Plot

```python
fig = pyplot.figure(figsize = (12,15))
ax = fig.add_subplot(111)
plt.ion()

fig.show()
fig.canvas.draw()

#iterate - plot, erase
for iterator in range(0,40):
    #print i
    ax.clear() # - Clear
    histogram_bins = sorted(randoms[iterator])
    fit = stats.norm.pdf(histogram_bins , np.mean(histogram_bins) , np.std(histogram_bins))
    ax.plot(histogram_bins , fit , '-o') # fit the line
    ax.hist(histogram_bins , normed = True) # fit the histogram
    fig.canvas.draw()   # draw
    time.sleep(0.5)    #sleep
```
