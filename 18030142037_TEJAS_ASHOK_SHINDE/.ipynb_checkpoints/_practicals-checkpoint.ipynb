```python
import pandas
import sqlite3
```

## Establishing Database Connection

```python
conn = sqlite3.connect('./../Datasets/coupons.db')
```

```python
merchant_DF = pandas.read_sql('select * from merchant' , conn)
customer_DF = pandas.read_sql('select * from customer' , conn)
coupon_DF = pandas.read_sql('select * from coupon' , conn)
sales_DF = pandas.read_sql('select * from sale' , conn)
```

```python
coupon_DF = coupon_DF.rename(columns = {'name' : 'coupon_name'})
merchant_DF = merchant_DF.rename(columns = {'name' : 'merchant_name'})
customer_DF = customer_DF.rename(columns = {'name' : 'customer_name'}) 
```

```python
sales_DF.ts = pandas.to_datetime(sales_DF.ts)
merge_CS = pandas.merge(customer_DF , sales_DF , on = 'id' , how = 'inner')
merge_CS = pandas.merge(merge_CS , coupon_DF , on = 'id' , how = 'inner')
merge_CS = pandas.merge(merge_CS , merchant_DF , on = 'id' , how = 'inner')
```

## 1. Top 10 customers who spent maximum amount (consider discounted price) in month of March.

```python
# merge_CS.sort_values('bill_amount' , ascending = False).head(10)
merge_CS['discount_amount'] = (merge_CS.bill_amount * 3.5) / 100
merge_CS.discount_amount.apply(lambda x : 300 if x > 300 else x)
merge_CS['discounted_rate'] = merge_CS.bill_amount - ((merge_CS.bill_amount * 3.5) / 100)
# merge_CS.sort_values('discounted_rate' , ascending = False).head(10)
merge_CS = merge_CS.set_index('ts')
```

```python
merge_CS[merge_CS.index.month == 3].sort_values('discounted_rate' , ascending = False).head(10)
```

## 2. Top 20 merchants with maximum turnover (complete data) on the e-commerce platform

```python
pandas.merge(merchant_DF , sales_DF , on = 'id' , how = 'outer').groupby('merchant_name').sum().sort_values('bill_amount' , ascending = False).head(20)
```

## 3. Top 10 coupons with maximum utilization

```python
cMerge = pandas.merge(coupon_DF , sales_DF , on = 'id').groupby('coupon_id').count()
cMerge.sort_values('coupon_name' , ascending = False).head(10)
```

## 4. Top 3 coupons which were utilized in each month

```python
# merge_CS['month_number']
# merge_CS.index.month
merge_CS['month_number'] = merge_CS.index.month
merge_pivot = merge_CS.pivot(index = 'id' , columns = 'month_number' , values = 'coupon_id').fillna(0)
merge_pivot.apply(lambda x : sorted(x)).tail(3)
```

## 5. Top 50 coupons for which e-commerce platform gave discount in terms of money spent.

```python
merge_MC = pandas.merge(coupon_DF , merchant_DF , how = 'inner' , on = 'id')
merge_MC = pandas.merge(merge_MC , sales_DF , how = 'inner' , on = 'id')
# merge_MC
merge_MC['merchant_paid'] = (merge_MC.bill_amount * merge_MC.discount)/100
# names = merge_MC[['merchant_id' , 'merchant_name']]
# names = names.set_index('merchant_id')
merge_MC[['merchant_id','merchant_paid']].groupby('merchant_id').sum().sort_values('merchant_paid' , ascending = False).head(50)
```

## 6. Plot graph for number of coupons used in each month

```python
# merge_CouponsMonth = 
# coupon_DF
merge_CouponsMonth = pandas.merge(coupon_DF,sales_DF , how = 'inner' , on = 'id')
merge_CouponsMonth = merge_CouponsMonth.set_index('ts')
```

```python
merge_CouponsMonth['month_number'] = merge_CouponsMonth.index.month
# merge_CouponsMonth.groupby('month_number').count()
# merge_CouponsMonth
merge_CouponsMonth[['coupon_id','month_number']].groupby('month_number').count().plot(kind = 'bar' , figsize = (15,8))
```
