```python
%matplotlib inline
import pandas
from matplotlib import pyplot as pyplot
from matplotlib import style
import random
import numpy
```

```python
figure = pyplot.figure(figsize = (20,14))
style.use('bmh')
def createPlotPoints():
    x = []
    y = []
    
    for iterator in range(10):
        xs = iterator
        ys = random.randrange(10)
        
        x.append(xs)
        y.append(ys)
        
    return x , y

aPlot = figure.add_subplot(2,2,1)
bPlot = figure.add_subplot(2,2,2)
cPlot = figure.add_subplot(2,2,(3,4))

x , y = createPlotPoints()
aPlot.plot(x , y , color = 'r' , label = "Test")
aPlot.legend()

x , y = createPlotPoints()
bPlot.plot(x , y)

x , y = createPlotPoints()
cPlot.plot(x , y)


pyplot.show()
```

```python
def createStudentMarks():
    return numpy.random.randint(0,100,5) , numpy.random.randint(0,100,5) , numpy.random.randint(0,100,5) , numpy.random.randint(0,100,5)

base = [1 , 2 ,3 , 4 , 5]
a , b , c , d = createStudentMarks()

pyplot.figure(figsize = (20,15))
# pyplot.plot([] , [] , color = 'y' , label = "Base" , linewidth = 5)
# pyplot.plot([] , [] , color = 'r' , label = 'Tejas' , linewidth = 5)
# pyplot.plot([] , [] , color = 'c' , label = 'Aditya' , linewidth = 5)
# pyplot.plot([] , [] , color = 'g' , label = 'Prathamesh' , linewidth = 5)
# pyplot.plot([] , [] , color = 'b' , label = 'Danish' , linewidth = 5)


pyplot.stackplot(base , a.tolist() , b.tolist() , c.tolist() , d.tolist() , colors = ['r' , 'c' , 'g' , 'b'])
pyplot.legend(["Adity" , "Tejas" , "Sarthak" , "Danny"])
pyplot.title('Overview of marks of students using stackplot')
pyplot.xlabel("Subject wise marks")
pyplot.ylabel("Stack of marks")
```
