```python
from matplotlib import pyplot as plt
from numpy import arange
x = arange(1,100)

# Using a sample y = mx + c linear regression formula where m, i.e slope is 2 and c is 5
y = 2 * x + 5
plt.title("Linear Regression plt.")
plt.xlabel("Experience in terms of Percentage")
plt.ylabel("Years of Experience")
plt.plot(x,y)
plt.show
```

```python
x = arange(1,1000)
y = 2 * x + 3

plt.title("Dashed Line")
plt.xlabel("Experience in terms of Percentage")
plt.ylabel("Years of Experience")
plt.plot(x,y,"--r")
plt.show
```

```python
import numpy as np 
import matplotlib.pyplot as plt  
   
# Compute the x and y coordinates 
x = np.arange(0, 3 * np.pi, 0.1) 
y_sin = np.sin(x) 
y_cos = np.cos(x)  
   
# Set up a subplot grid that has height 2 and width 1, 
# and set the first such subplot as active. 
plt.subplot(2, 1, 1)
   
# Make the first plot 
plt.plot(x, y_sin) 
plt.title('Sin')  
   
# Set the second subplot as active, and make the second plot. 
plt.subplot(2, 1, 2) 
plt.plot(x, y_cos) 
plt.title('Cosin')  
   
# Show the figure. 
plt.show()
```
