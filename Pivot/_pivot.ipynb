```python
dataframe = pandas.read_csv('outfile.csv' , parse_dates = ['TimeStamp'])
# dataframe.dtypes
dataframe
```

```python
df = dataframe[dataframe['Status'] == 'Ban'][['IP','Status']].groupby('IP').count()
df.reset_index(inplace = True)
```

```python
reader = geoip2.database.Reader('../Datasets/GeoLite2-Country_20190101/GeoLite2-Country.mmdb')
```

```python
df['Country'] = df.IP.map(lambda x : reader.country(x).country.name)
```

```python
df.groupby('Country').sum().sort_values('Status' , ascending = False).head(10).plot(kind = 'bar')
```
