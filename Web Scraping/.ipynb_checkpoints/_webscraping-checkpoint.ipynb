```python
# IMPORTS
import bs4
from urllib.request import urlopen as urlRequest
from bs4 import BeautifulSoup as soup
import csv

'''
FILE STREAM
'''
filename = "productData.csv"
dataFile = open("./productData.csv" , "w")
headers = "Title , Brand , Actual Price , Deal price \n"
dataFile.write(headers)

'''
SCRIPT
'''
url = "https://www.amazon.in/b/ref=s9_acss_bw_cg_ps4ctf_1b1_w?node=2591141031&pf_rd_m=A1K21FY43GMZF8&pf_rd_s=merchandised-search-3&pf_rd_r=8VB47NJWQ7GYGQG6FYG2&pf_rd_t=101&pf_rd_p=cb6a9007-83fc-4d45-9106-4f68de423aa6&pf_rd_i=2591138031"

# Opens a connection to grab a page
urlClient = urlRequest(url)

# Offloads the content in a variable
htmlPage = urlClient.read()

# Closes the client
urlClient.close()

# HTML Parsing
pageSoup = soup(htmlPage,"html.parser").encode("utf-8")

# Grab each item and store in variable items
items = pageSoup.findAll("li",{"class":"a-carousel-card acswidget-carousel__card"})


# item = items[0]
# print(item.div.a["title"])

'''
PUTTING SCRAPED DATA INTO THE CSV FILE
'''
try:    
    for item in items:

        title = item.div.a["title"]

        if not item.findAll("div",{"class":"a-box-group acs_product-metadata__brand"}) :
            brand = "NULL"
        else :
            brand = item.findAll("div",{"class":"a-box-group acs_product-metadata__brand"})[0].text

        if not item.findAll("span",{"class":"a-size-mini a-color-secondary acs_product-price__list a-text-strike"}) :
            actualPrice = "0"
        else :    
            actualPrice = item.findAll("span",{"class":"a-size-mini a-color-secondary acs_product-price__list a-text-strike"})[0].text.split()[1]

        if not item.findAll("span",{"class" : "a-size-base a-color-price acs_product-price__buying"}) :
            dealPrice = 0
        else :     
            dealPrice = item.findAll("span",{"class" : "a-size-base a-color-price acs_product-price__buying"})[0].text.split()[1]

#         dataFile.write(title.replace(",","") + "," + brand.replace(",","") + "," + actualPrice.replace(",","") + "," + dealPrice.replace(",","") + "\n")
        print(title)
        print(brand)
        print(actualPrice)
        print(dealPrice)
except FileNotFoundError :
    print("File not found in storage.")
finally:    
    dataFile.close()   
```
