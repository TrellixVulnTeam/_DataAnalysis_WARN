## Web Scraping Data from Twitter in accordance with the input query

```python
# IMPORTS
import bs4
from urllib.request import urlopen as urlRequest
from bs4 import BeautifulSoup as soup
import csv

'''
FILE HANDLER
'''
# fileOut = open("./tmp/twitterData.csv" , "wb")
# writer = csv.writer(fileOut)

# writer.writerow(headers)
headers = ["HANDLE" , "NAME" , "TWEET" , "LINK"]

fileOut = open('./tmp/twitterData.csv' , 'w')
writer = csv.writer(fileOut)
writer.writerow(headers)

'''
PAGE SOUP AND SCRAPING THE HTMLETC
'''
# URL client initialization and reading site data
query = 'trump'
url = 'https://mobile.twitter.com/search?q=' + query
print(url)
# Opens a connection to grab a page
urlClient = urlRequest(url)
# Offloads the content in a variable
htmlPage = urlClient.read()
# Closes the client
urlClient.close()
# HTML Parsing
pageSoup = soup(htmlPage,"html.parser")
# Tweet Detail Fetch
tweetStructure = pageSoup.findAll("table" , {"class" : "tweet"})


'''
REGEX's
'''
twitterHandleRegex = "^[/]([^?]*)"
preLink = "https://twitter.com/"


'''
EXTRACTING SCRAPED DATA
'''
# twitterHandle = findall(twitterHandleRegex,tweetStructure[0].tr.td.a['href'])[0]
# print(tweetStructure[0].findAll("strong" , {"class" : "fullname"})[0].text)
# print(preLink + tweetStructure[0].tr.td.a['href'])
# tweet = tweetStructure[0].findAll("div" , {"class" : "tweet-text"})[0].text
# print(tweet)

try:
    for tweet in tweetStructure:
        struct = []
        
        twitterHandle = tweet.findAll("div" , {"class" : "username"})[0].text
        struct.append(twitterHandle.strip())
        
        fullName = tweet.findAll("strong" , {"class" : "fullname"})[0].text
        struct.append(fullName.strip())
        
        tweetText = tweet.findAll("div" , {"class" : "tweet-text"})[0].text
        struct.append(tweetText.strip())
        
        accountLink = preLink + twitterHandle[1:]
        struct.append(accountLink.strip())
        print(accountLink)
        
#         print(struct)
#         print("\n\n")
#         print(twitterHandle + "," + fullName + "," + accountLink + "," + tweetText + "\n")
#         fileOut.write(twitterHandle + "," + fullName + "," + accountLink + "," + tweetText + "\n")
#         print(struct)
        if writer.writerow(struct):
            print("Success")
#         print("Handle : " + twitterHandle + " Name : " + fullName + " Account Link : " + accountLink + "\nTweet : " + tweetText)
#         print(fullName)
except FileNotFoundError:
    print("Something went wrong while opening the file")
finally:
    fileOut.close()
    
```
