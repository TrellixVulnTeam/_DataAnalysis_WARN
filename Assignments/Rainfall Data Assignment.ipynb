### IMPORTS

```python
import pandas
import matplotlib.pyplot as pyplot
import numpy
```

### Reading the file into a dataframe

```python
location = './../Datasets/datafile.csv'
dataframe = pandas.read_csv(location)
```

## 1. Pie chart which shows rainfall for all 12 Months

```python
# User Defined Percentage

# pyplot.pie(
#               (numpy.sum(dataframe.iloc[: , 2:14]) / numpy.sum(numpy.sum(dataframe.iloc[: , 2:14])) * 100),
#                labels = dataframe.columns[2:14],
# #                autopct='%1.1f%%',
#           )


# Automatically created Percentage
pyplot.style.use('classic')
pyplot.figure(figsize = (15,10))
pyplot.title('Rainfall for all 12 months' , pad = 50 , fontdict = {'size' : 22})
pyplot.pie(
              (numpy.sum(dataframe.iloc[: , 2:14])),
               labels = dataframe.columns[2:14],
               autopct = '%1.1f%%',
               colors = pyplot.cm.prism(numpy.linspace(.1, .5, len(dataframe.columns[2:14]))) 
          )
pyplot.legend(bbox_to_anchor=(1.1, 1.05))
pyplot.axis('equal')
```

## 2. Bar graph for annual rainfall for each state

```python
pyplot.style.use('ggplot')
pyplot.figure(figsize = (15,10))
pyplot.xticks(rotation = 90)
pyplot.xlabel('States / Union Territories')
pyplot.ylabel('Annual Rainfall')
pyplot.title('Bar graph for annual rainfall for each state')
pyplot.bar(
           dataframe.groupby(dataframe['STATE/UT']).sum().index , 
           dataframe.groupby(dataframe['STATE/UT']).sum().ANNUAL,
           color = 'c'
          )
```

## 3. Line graph where months are on X-axis, rainfall for each month on Y-axis. Plot 5 states with lowest annual rainfall.

### 3a. Line graph where months are on X-axis, rainfall for each month on Y-axis

```python
pyplot.style.use('seaborn')
pyplot.figure(figsize = (15,10))
pyplot.title('Line graph where months are on X-axis, rainfall for each month on Y-axis')
pyplot.xlabel('Months')
pyplot.ylabel('Rainfall')
pyplot.plot(dataframe.columns[2:14] , numpy.sum(dataframe.iloc[: , 2:14]))
```

### 3b. Plotting 5 states with lowest annual rainfall

```python
# dataframe.groupby('STATE/UT').sum().sort_values('ANNUAL').head(5)
pyplot.figure(figsize = (15,10))
pyplot.style.use('fivethirtyeight')
pyplot.subplot(1,2,1)
pyplot.title('A bar graph for the 5 States/UT with lowest rainfall')
pyplot.xlabel('States/UT')
pyplot.ylabel('Rainfall')
pyplot.xticks(rotation = 90)
pyplot.bar(
            dataframe.groupby('STATE/UT').sum().sort_values('ANNUAL').head(5).index,
            dataframe.groupby('STATE/UT').sum().sort_values('ANNUAL').head(5).ANNUAL
           )

pyplot.subplot(1,2,2)
pyplot.title('A line graph for the 5 States/UT with lowest rainfall')
pyplot.xlabel('States/UT')
pyplot.ylabel('Rainfall')
pyplot.xticks(rotation = 90)
pyplot.plot(
            dataframe.groupby('STATE/UT').sum().sort_values('ANNUAL').head(5).index,
            dataframe.groupby('STATE/UT').sum().sort_values('ANNUAL').head(5).ANNUAL,
            color = 'g'
           )
```

## 4. Line graph where months are on X-axis and median of rainfall for each month on Y-axis.

```python
pyplot.figure(figsize = (15,10))
pyplot.style.use('dark_background')
pyplot.title('Line graph for median of rainfall for each month')
pyplot.xlabel('Months')
pyplot.ylabel('Medians')
pyplot.plot(dataframe.columns[2:14] , numpy.median(dataframe.iloc[: , 2:14] , axis = 0))
```
